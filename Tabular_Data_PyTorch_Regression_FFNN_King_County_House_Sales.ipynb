## 7. Train_Test_Split

```python
y = df.price.values.reshape(-1, 1)
X = df.drop(columns=['price']).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## 8. Scale Data

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## 9. Build Model

```python
class Model(nn.Module):
  def __init__(self, in_features=X_train.shape[1]):
    super().__init__()
    self.layers = nn.Sequential(
      nn.Linear(in_features, 64),
      nn.ReLU(),
      nn.Linear(64, 64),
      nn.ReLU(),
      nn.Linear(64, 1),
      )

  def forward(self, x):
    y_pred = self.layers(x)
    return y_pred
```

```python
model = Model()
model.to(device)
```

```python
summary(model, input_size=(X_train.shape))
```

```python
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters())
```

## 10. Train Model

```python
class Dataset(Dataset):

  def __init__(self, X, y):
    self.data = torch.Tensor(X)
    self.label = torch.Tensor(y)
    self.len_data = len(self.data)

  def __getitem__(self, item):
    return self.data[item], self.label[item]

  def __len__(self):
    return self.len_data
```

```python
batch_size = 32

dataset_train = Dataset(X_train, y_train)
train_loader = DataLoader(dataset=dataset_train,
                          batch_size=batch_size,
                          shuffle=True)

dataset_test = Dataset(X_test, y_test)
test_loader = DataLoader(dataset=dataset_test,
                          batch_size=batch_size,
                          shuffle=False)
```

```python
n_epochs = 100
train_losses = np.zeros(n_epochs)
test_losses = np.zeros(n_epochs)

for i in range(n_epochs):

  model.train()
  train_loss = []
  for inputs, targets in train_loader:
    inputs, targets = inputs.to(device), targets.to(device)
    optimizer.zero_grad()
    outputs_train = model(inputs)
    loss_train = criterion(outputs_train, targets)

    loss_train.backward()
    optimizer.step()

    train_loss.append(loss_train.item())

  train_loss = np.mean(train_loss)


  model.eval()
  test_loss = []
  for inputs, targets in test_loader:
    inputs, targets = inputs.to(device), targets.to(device)
    outputs_train = model(inputs)
    loss_test = criterion(outputs_train, targets)

    test_loss.append(loss_test.item())

  test_loss = np.mean(test_loss)

  train_losses[i] = train_loss
  test_losses[i] = test_loss

  if (i+1) % 10 == 0:
    print(f'Epoch {i+1}/{n_epochs}, Train Loss: {loss_train.item():.4f}, Test Loss: {loss_test.item():.4f}')
```

```python
plt.plot(train_losses, label='train loss')
plt.plot(test_losses, label='test loss')
plt.legend()
```

## 11. Evaluate

```python
with torch.no_grad():
  y_val_tensor = model.forward(torch.Tensor(X_test))
  y_val = y_val_tensor.numpy()
```

```python
np.sqrt(mean_squared_error(y_test, y_val))
```

```python
mean_absolute_error(y_test, y_val)
```

```python
explained_variance_score(y_test, y_val)
```
